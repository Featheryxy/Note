```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np 
```

```python
x = np.linspace(0, 10, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y, '--r')
plt.plot(x, cos_y, 'ob')
plt.xlim(-5, 15)
plt.ylim(0, 1.5)
plt.axis('tight')
plt.show()
```

```python
x = np.linspace(0, 10, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)
plt.plot(x, sin_y, '--r', label='sinx')
plt.plot(x, cos_y, 'ob', label='cosx')
plt.axis('tight')
plt.legend()
plt.title('sinx and cosx')
plt.show()
```

### Scatter Plot

```python
plt.scatter(x, np.sin(x))
plt.show()
```
