```python
import os
import re
import cv2
import numpy as np
```

```python
X_0 = []
X_1 = []
path = 'liewen'

# X_0_re匹配文件中无裂纹的图片， X_1_re匹配文件中有裂纹的图片
X_0_re = r'\d*\_\d*\.jpg'
X_1_re = r'\d*\_\d*\_3\.jpg'
# 只选取160 * 160 pixel 的文件夹下
match = r'liewen\\lw\d*\\160'

# 将有裂纹的图片名称保存到X_1列表中
# 将无裂纹的图片名称保存到X_0列表中
# 读取path底下所有的160*160 pixel的图片，
for root, dirs, files in os.walk(path):
    if re.match(match, root):
        for fil in files:
            path_name = root + os.sep + fil
            print(path_name)
            if re.match(X_0_re, fil):
                X_0.append(path_name)
            elif re.match(X_1_re, fil):
                X_1.append(path_name)       
```

```python
# 将关键点描述组成输入矩阵X
def get_data(X_0):
    X_fu = []
    for img in X_0:
        # 读取图片
        img = cv2.imread(img, 0)
        # 创建sift对象
        sift = cv2.xfeatures2d.SIFT_create()
        # kp 为关键点， des为关键点的描述。 大小为 关键点的数量 * 128
        kp, des = sift.detectAndCompute(img, None)
        # 剔除图片中没有关键点的 des
        if np.all(des==None):
            continue
        # 剔除图片中没有关键点数量小于10 的 des
        if len(des) < 10:
            continue
        # 提取一张图片的前十个关键点描述， 
        #  重组矩阵，使其一行表示一张图片，列为 10 * 128
        x = des[:10].reshape((1, -1))
        X_fu.append(x[0])
        
    return X_fu
```

```python
# X_fu 保存没有裂纹的照片的关键点描述
X_fu = get_data(X_0)
X_fu = np.array(X_fu)

# X_zheng 保存有裂纹的照片的关键点描述
X_zheng = get_data(X_1)
X_zheng = np.array(X_zheng)

# 将X_fu，X_zheng组成输入矩阵X
X = np.vstack((X_fu, X_zheng))
```

```python
y0 = np.array([0] * len(X_fu))
y1 = np.array([1] * len(X_zheng))
# y 为X矩阵中样本的标签， 0表示无裂纹， 1表示有裂纹
y = np.hstack((y0, y1))
```

```python
from sklearn.model_selection import train_test_split

# 将数据分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### 使用KNN进行分类

```python
from sklearn.neighbors import KNeighborsClassifier

# 生成knn对象， 距离使用明科夫斯基距离， p=3
clf = KNeighborsClassifier(n_neighbors=4, p=3, weights='distance')
clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)
print('分类器得分： ', clf.score(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_predict)
```

```python
from sklearn.metrics import precision_score

precision_score(y_test, y_predict)
```

```python
from sklearn.metrics import recall_score

recall_score(y_test, y_predict)
```

```python
from sklearn.metrics import f1_score

f1_score(y_test, y_predict)
```
