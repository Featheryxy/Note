```python
import os
import re
import cv2
import numpy as np
```

```python
X_0 = []
X_1 = []
path = 'liewen'

# X_0_re X_1_re
X_0_re = r'\d*\_\d*\.jpg'
X_1_re = r'\d*\_\d*\_3\.jpg'
# 160 * 160 pixel 
match = r'liewen\\lw\d*\\160'

# X_1
# X_0
# path160*160 pixel
for root, dirs, files in os.walk(path):
    if re.match(match, root):
        for fil in files:
            path_name = root + os.sep + fil
            print(path_name)
            if re.match(X_0_re, fil):
                X_0.append(path_name)
            elif re.match(X_1_re, fil):
                X_1.append(path_name)       
```

```python
# X
def get_data(X_0):
    X_fu = []
    for img in X_0:
        # 
        img = cv2.imread(img, 0)
        # sift
        sift = cv2.xfeatures2d.SIFT_create()
        # kp  des   * 128
        kp, des = sift.detectAndCompute(img, None)
        #  des
        if np.all(des==None):
            continue
        # 10  des
        if len(des) < 10:
            continue
        #  
        #   10 * 128
        x = des[:10].reshape((1, -1))
        X_fu.append(x[0])
        
    return X_fu
```

```python
# X_fu 
X_fu = get_data(X_0)
X_fu = np.array(X_fu)

# X_zheng 
X_zheng = get_data(X_1)
X_zheng = np.array(X_zheng)

# X_fuX_zhengX
X = np.vstack((X_fu, X_zheng))
```

```python
y0 = np.array([0] * len(X_fu))
y1 = np.array([1] * len(X_zheng))
# y X 0 1
y = np.hstack((y0, y1))
```

```python
from sklearn.model_selection import train_test_split

# 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier

# knn  p=3
clf = KNeighborsClassifier(n_neighbors=4, p=3, weights='distance')
clf.fit(X_train, y_train)
y_predict = clf.predict(X_test)
print(' ', clf.score(X_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_predict)
```

```python
from sklearn.metrics import precision_score

precision_score(y_test, y_predict)
```

```python
from sklearn.metrics import recall_score

recall_score(y_test, y_predict)
```

```python
from sklearn.metrics import f1_score

f1_score(y_test, y_predict)
```
